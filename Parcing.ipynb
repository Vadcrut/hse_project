Cianparser - это библиотека для парсинга сайта Циан. С его помощью можно получить достаточно подробные и структурированные данные по краткосрочной и долгосрочной аренде, продаже квартир, домов, танхаусов итд.

```python
import cianparser
```

```python
moscow_parser = cianparser.CianParser(location="Москва") #смотрим по городу: Москва
data = moscow_parser.get_flats(rooms=1,deal_type="sale", with_saving_csv=True, additional_settings={"start_page":1, "end_page":54})
```
