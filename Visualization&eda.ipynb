```python
#уверенно импортируем библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('flat_with_1_room.csv',delimiter=';') #считываем фай
df.head()
```

```python
df.shape
```

```python
df2=pd.read_csv('flat_with_2_rooms.csv',delimiter=';') #считываем файл, где 2 комнаты
df2.head()
```

```python
df2.shape
```

```python
df3=pd.read_csv('flats_with_3_rooms.csv',delimiter=';') #считываем файл, где 3 комнаты
df2.head()
```

```python
data=pd.concat([df, df2,df3])
data.shape
```

<font color="blue">**Описание признаков:**</font>

1. **author** - автор объявления.
2. **author_type**	 - тип автора (ex.: real_estate_agent - агентство недвижимости<br>
                                    homeowner - собственник<br>
                                    realtor - риелтор<br>
                                    official_representative - ук оф.представитель<br>
                                    representative_developer - представитель застройщика<br>
                                    developer - застройщик)<br>
3. **url** - ссылка объвления.
4. **location** - локация(город) объявления, в нашем случае Москва.
5. **deal_type** - тип объявления, к примеру, долгосрочная, краткосрочная аренда, продажа ("rent_long", "rent_short", "sale").
6. **accommodation_type** -тип аппартаментов (квартира, дом, танхаус и т.д.), в нашем случае квартира.
7. **floor** - этаж, накотором расположена квартира
8. **floors_count** - общее количество этажей в доме
9. **rooms_count** - количество комнат, в нашем случае 1 комната в одном файле и 2 в другом.
10. **total_meters** -общая площадь квартиры.
11. **price** -цена квартиры.
12. **district** - название района.
13. **street** - название улицы.
14. **house_number** - номер дома.
15. **underground** - название метро.
16. **residential_complex** - название комплекса/ЖК.

Теперь нужно преобразовать таблицу, убрав лишнее.

```python
cols=['Этаж','Общее количество этаже в доме','Количество комнат','Площадь квартиры','Цена','Район','Улица','Метро']
```

Колонки, которые мы будем исследовать, нужно убрать остальные, так как они нам либо не дают нужную информацию, либо совпадают у всех объявлений.

```python
data = data.drop(['author','author_type','location','accommodation_type','house_number','residential_complex','url'],axis=1)
```

```python
data['deal_type'].unique()
```

Так же совпадает у всех объявлений-убираем.

```python
data=data.drop('deal_type',axis=1)
```

```python
data.columns=cols
```

```python
data.head()
```

```python
data=data.dropna()
```

```python
data.describe()
```

Используем метод corr с парметром numeric_only=True, чтобы узнать как числовые переменные коррелируют друг с другом

```python
data.corr(numeric_only=True)
```

Сделаем матрицу корреляции:

```python
sns.heatmap(data.corr(numeric_only=True), annot=True, fmt = '0.1f', vmin=-1, vmax=1, center= 0, cmap= 'coolwarm')
plt.show()
```

По данной матрице мы можем увидеть некоторые закономерности:
1. Площадь квартиры сильно коррелирует с ее ценой.
2. Этаж не сильно влияет на цену.
3. Отрицательная корреляция между общим количество этажей и ценой.
4. Количество комнат коррелирует с ее ценой.


**Зависимость цены от этажа (линейный график)**

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Этаж', y='Цена', data=data)
plt.title('Зависимость цены от этажа')
plt.xlabel('Этаж')
plt.ylabel('Цена (в рублях)')
plt.show()
```

Этот график показывает зависимость цены от этажа. Можно заметить, что цена начинает резко расти на очень высоких этажах, что может свидетельствовать о том, что скорее всего это квартиры в новых домах, где есть больше 50 этажей. Давайте приблизим график и посмотретрим эту зависимость в квартирах до 30ого этажа.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Этаж', y='Цена', data=data)
plt.title('Зависимость цены от этажа')
plt.xlim(0,30)
plt.xlabel('Этаж')
plt.ylabel('Цена (в рублях)')
plt.show()
```

Можем заметить скачок в цене на примерно 12 и 26 этажах, что может свидетельствовать, что квартиры на средних этажах дороже квартир на первых и последних.

**Зависимость площади квартиры от района**

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Район'], data['Площадь квартиры'])
plt.xlim(0, 50)
plt.xticks(rotation=90) # Поворот названий районов для лучшей читаемости
plt.suptitle('Зависимость площади квартиры от района')
plt.xlabel('Район')
plt.ylabel('Площадь, м.кв.')
plt.show()
```

**Зависмость цены квартиры от района**

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Район'], data['Цена'],color='pink')
plt.xlim(0, 50)
plt.xticks(rotation=90)
plt.suptitle('Зависимость цены квартиры от района')
plt.xlabel('Район')
plt.ylabel('Цена, руб.')
plt.show()
```

Boxplot цен по районам

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Район', y='Цена', data=data)
plt.title('Распределение цен по районам')
plt.xlabel('Район')
plt.ylabel('Цена (в рублях)')
plt.xticks(rotation=90, ha='right')
plt.xlim(0, 25)
plt.tight_layout() #Чтобы надписи не накладывались друг на друга
plt.show()
```

Boxplot позволяет сравнить распределение цен в разных районах. Он показывает медиану (линия внутри прямоугольника), квартили (границы прямоугольника) и выбросы (точки за пределами "усов"). Это помогает определить, в каких районах квартиры в среднем дороже, и в каких есть значительные отклонения от среднего. В пресненском районе замечены сильные выбросы, что может говорить о неоднородности квартир в нем, такие же выбросы есть в Замоскворечье и Басманном.

**Распределение цен**

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Цена'], color = 'purple', fill = True)
plt.suptitle('Распределение цены')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.show()
```

Из графика можно заметить, что распределение цены напоминает нормальное, но мода, математическое ожидание и медиана немного смещены. Это может говорить о будущем росте цен на квартиры.
Давайте найдем среднее цены:

```python
data['Цена'].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Район', data=data, color = 'orange')
plt.suptitle('Распределение по районам')
plt.xlabel('Район')
plt.ylabel('Частота')
plt.xticks(rotation=90)
plt.xlim(0, 50)
plt.show()
```

Можем заметить, что Пресненский район намного популярнее отсальных, что можем свидетельствовать о количестве новостроек и популярности района.

```python
data=pd.read_csv('/content/data.csv', delimiter=',')
data.head()
```

```python
data = data.dropna()
```

```python
data['Расстояние до метро']
```

```python
df_filtered = data[['Цена', 'Расстояние до метро']].dropna()

# Конвертация цены в миллионы рублей для удобства отображения
df_filtered['Цена'] = df_filtered['Цена'] / 1e6

# Построение графика
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='Расстояние до метро', y='Цена', alpha=0.5)
sns.regplot(data=df_filtered, x='Расстояние до метро', y='Цена', scatter=False, color='red')
plt.xlabel("Расстояние до метро (м)")
plt.ylabel("Цена квартиры (млн руб.)")
plt.title("Зависимость цены квартиры от расстояния до метро")
plt.grid(True)
plt.show()
```

Легко заметить, что с увелечинем удаленности от метро, сила влияния этого фактора на цену квартиры снижается. Что также подтверждается линией тренда
