```python
import requests

API_KEY = "4196ac17-3d16-4d03-aebf-cc345bcdf6c1"
```

```python
def get_point(ADDRESS, metro):
  url = f"https://geocode-maps.yandex.ru/1.x/?apikey={API_KEY}&geocode={ADDRESS}&format=json"
  response = requests.get(url)
  data = response.json()

  pos = data["response"]["GeoObjectCollection"]["featureMember"][0]["GeoObject"]["Point"]["pos"]
  longitude, latitude = map(float, pos.split())

  METRO_NAME = f'{metro}, Москва'

  url_m = f"https://geocode-maps.yandex.ru/1.x/?apikey={API_KEY}&geocode=метро+{METRO_NAME}&kind=metro&format=json"
  response_m = requests.get(url_m)
  data_m = response_m.json()

  # Извлекаем координаты
  pos_m = data_m["response"]["GeoObjectCollection"]["featureMember"][0]["GeoObject"]["Point"]["pos"]
  longitude_m, latitude_m = map(float, pos_m.split())

  return haversine(latitude, longitude, latitude_m, longitude_m)


```

```python
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371 * 1000  # Радиус Земли в метрах
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c
```

```python
distances = []
for i, j in data.iterrows():
  address = f'Москва, {j["Район"]}, {j["Улица"]}, {j["Номер дома"]}'
  distances.append(get_point(address, j['Метро']))
```
